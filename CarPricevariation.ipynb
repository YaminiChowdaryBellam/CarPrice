## Hypothesis Testing

For Hypothesis testing we perform the below Tests

Test if price or log(price) follows normal distribution

Test for significance of price/log(price) stratified by:
a. Fuel type (diesel Vs gas)
b. Aspiration (Turbo Vs Standard)

Apply ANOVA to compare the price (log(price)) of autos stratified by Body Style.

```r
#Compare & test normality of the distribution of price & log(price).

sample1 <- data_sub$price

# random sample from normal distribution with  mean(price) and standard deviation = sd(price)
Mean <- mean(sample1)
SD <- sd(sample1)
cat(paste('Mean of auto price = ', as.character(round(Mean,2)),', variance = ', as.character(round(SD^2,2)),', 
            standard deviation = ', as.character(round(SD,2))))
sample2 <- rnorm(195, mean = Mean, sd = SD)

sampleLog1 <- log(data_sub$price)

# random sample from normal distribution with  mean(log(price)) and standard deviation = sd(log(price))

lm <- mean(sampleLog1)
ls <- sd(sampleLog1)
cat(paste('\nMean of logarithm of auto price = ', as.character(round(lm,2)),', variance = ', as.character(round(ls^2,2)),', 
         standard deviation = ', as.character(round(ls,2))))
sampleNorm2 <- rnorm(195, mean = lm, sd = ls)

#Plot the density curve of the price variable
plot1 <- ggplot(data_sub, aes(price)) + geom_density()

#Plot the density curve of the log(price) variable
plot2 <- ggplot(data_sub, aes(log(price))) + geom_density()

grid.arrange(plot1, plot2, ncol = 2)
```

Here, we can observe that that distribution is skewed.

```r
#Test for significance of price/log(price) stratified by:
#a. Fuel type (diesel Vs gas)
#b. Aspiration (Turbo Vs Standard)

#Let's split the price vector by Fuel Type
table(data_sub$fuel.type)
priceFType <- split(data_sub$price, as.factor(data_sub$fuel.type), drop = FALSE)

#Compute logarithms
logPriceGas <- log(priceFType$gas)
logPriceDiesel <- log(priceFType$diesel)

#Let's split the price vector by Aspiration
table(data_sub$aspiration)
priceAspiration <- split(data_sub$price, as.factor(data_sub$aspiration), drop = FALSE)

#Compute logarithms
logPriceTurbo <- log(priceAspiration$turbo)
logPriceStandard <- log(priceAspiration$std)
```

Now we want to test if:

Price of gas cars is significantly different from that of diesel cars
Price of turbo aspiration cars are significantly different from that of std aspiration cars
For these tests, we shall use the logarithm of prices, since they more or less closer to being normally distributed than just the prices.

```r
#Testing if price of diesel cars vary signifcantly from that of gas cars

#Plot for comparing means
plot.t <- function(a, b, cols = c('title1', 'title2'), nbins = 20)
{
  maxs <- max(c(max(a), max(b)))
  mins <- min(c(min(a), min(b)))
  breaks = seq(maxs, mins, length.out = (nbins + 1))
  par(mfrow = c(2, 1))
  hist(a, breaks = breaks, main = paste('Histogram of', cols[1]), xlab = cols[1])
  abline(v = mean(a), lwd = 4, col = 'red') #Put a line at the mean
  hist(b, breaks = breaks, main = paste('Histogram of', cols[2]), xlab = cols[2])
  abline(v = mean(b), lwd = 4, col = 'red')
  par(mfrow = c(1, 1))
}

options(repr.plot.width=5, repr.plot.height=5) ## Set the plot area.
plot.t(logPriceDiesel,logPriceGas, c('logPrice of Diesel cars', 'logPrice of Gas cars'))
```

The red line in the graph shows the mean of the logarithm of prices.we can see that there is a small difference in means. Let's do a t-test to see if this is significant enough.

```r
t.test(logPriceDiesel,logPriceGas, alternative = "two.sided")
```

Since p-value is a little over 0.05 and 0 lies in the 95% confidence interval (-0.007204235, 0.470292829), we do not have sufficient evidence to reject the null hypothesis that the prices of diesel cars are not significantly different from that of gas cars.

```r
Price of turbo aspiration cars is significantly different from that of std aspiration cars
null hypothesis is The difference in average log(Price) of turbo aspiration cars and std aspiration cars is 0
The difference in average log(Price) is not equal to 0
```

```r

#Testing if price of turbo cars vary signifcantly from that of standard cars
options(repr.plot.width=5, repr.plot.height=5) ## Set the plot area.
plot.t(logPriceTurbo,logPriceStandard, c('logPrice of Turbo cars', 'logPrice of Standard cars'))
t.test(logPriceTurbo,logPriceStandard, alternative = "two.sided")
```

From the graphs, we can see there is some difference between the average prices

Since p-value is significantly less than 0.05 and 0 does not lie in the 95% confidence interval (0.1876238 0.4901126), we reject the null hypothesis that the prices of turbo cars are not significantly different from that of standard cars.This indicates that the prices of turbo cars are significantly different from that of standard cars.

```r
#Apply ANOVA to compare the price (log(price)) of autos stratified by Body Style. 
#Let's split the price vector by Body Style and display the results
table(data_sub$body.style)
priceBodyStyle <- split(data_sub$price, as.factor(data_sub$body.style), drop = TRUE)

#Note: There are 5 body styles. We are not considering hardtops & convertibles since there are very few observations on these
#Compute the logarithms
logPriceHatchback <- log(priceBodyStyle$hatchback)
logPriceSedan <- log(priceBodyStyle$sedan)
logPriceWagon <- log(priceBodyStyle$wagon)
```

```r
#Create a dataframe with the 3 groups as factors
df <- data.frame('group'=c(rep(1,length(priceBodyStyle$hatchback)),rep(2,length(priceBodyStyle$sedan)),rep(3,length(priceBodyStyle$wagon))),
                'val' = c(log(priceBodyStyle$hatchback),log(priceBodyStyle$sedan),log(priceBodyStyle$wagon)))
df$group <- factor(df$group) # Make sure your groups are a factor (for further analysis below)

#Plot as a boxplot
boxplot(df$val ~ df$group)

#Look at the basic properties of the new dataframe
str(df)
```

Box plot shows that group 1 may be different. so let's just run ANOVA test.

```r
df_aov <- aov(val ~ group, data = df) #Fischer's classic ANOVA function
summary(df_aov)
print(df_aov)
```

From the ANOVA table printed, we can see that there is a significant difference between at least a couple of the groups.  p-value is less than 0.05. From the box plot, we could guess it that group 1 is different. But we are not complete aware of what group 1 has.

```r
tukey_anova <- TukeyHSD(df_aov)  # Tukey's Range test:
tukey_anova
plot(tukey_anova)
```

By examining the plot above we can find pairs of groups that have a significant difference at the 95% confidence level. The confidence interval (shown by the length of the bar) will not overlap 0 when the difference is significant. We see that groups 2 & 1 (price of hatchback cars & price of sedans) differ significantly.

We found that the logarithm of price is more or less normally distributed
Using t-test, we found that:

 a. Price of gas cars is not significantly different from that of diesel cars
 b. Price of turbo aspiration cars is significantly different from that of std aspiration cars
Using ANOVA & Tukey's HSD we found that price of hatchback cars differs significantly from that of sedans, the price of hatchbacks and wagons may or may not differ, and the price of sedans and wagons do not differ significantly.
